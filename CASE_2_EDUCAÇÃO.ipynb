```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
Base_Dados = pd.read_csv('StudentsPerformance+(1).csv')

Base_Dados.shape
```

```python
Base_Dados.head()
```

```python
Nulos = Base_Dados.isnull()

plt.figure(figsize=(16,5))
plt.title('Analise dos Campos Nulos')
sns.heatmap(Nulos,cbar=False)
```

```python
plt.title("Analise de Outliers nas notas")
sns.boxplot(Base_Dados)
```

```python
plt.title('Analise  das Notas de Matematica por genero')
sns.boxplot(data=Base_Dados, x='math score', y='gender');
```

```python
plt.title('Analise  das Notas de Leitura por genero')
sns.boxplot(data=Base_Dados, x='reading score', y='gender');
```

```python
plt.title('Analise  das Notas de Escrita por genero')
sns.boxplot(data=Base_Dados, x='writing score', y='gender');
Base_Dados.groupby(by=['gender']).describe()['writing score'].reset_index()
```

```python
Base_Dados.groupby(by=['gender']).describe()['math score'].reset_index()
```

```python
sns.pairplot(data=Base_Dados, hue='race/ethnicity');
```

```python
plt.title('Analise  das Notas de Matematica por Raça')
sns.boxplot(data=Base_Dados, x='math score', y='race/ethnicity');
```

```python
plt.title('Analise  das Notas de Matematica por nivel de educação dos pais')
sns.boxplot(data=Base_Dados, x='math score', y='parental level of education');
Base_Dados.groupby(by=['parental level of education']).describe()['math score'].reset_index()
```

```python
plt.title('Analise  das Notas de Matematica por preparação por testes')
sns.boxplot(data=Base_Dados, x='math score', y='test preparation course');
Base_Dados.groupby(by=['test preparation course']).describe()['math score'].reset_index()
```

```python
plt.title('Analise de distibuição entre notas de Matematica e escrita')
sns.scatterplot( data=Base_Dados, x='math score', y='writing score');
```

```python
sns.jointplot( data=Base_Dados, x='math score', y='writing score');
```
