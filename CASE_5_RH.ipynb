```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
Base_Dados = pd.read_csv('Salary_Data.csv')

Base_Dados.head()
```

```python
Base_Dados.rename (columns={
    'YearsExperience' : 'Xp',
    'Salary' : 'Renda'

}, inplace=True)
```

```python
Base_Dados.shape
```

```python
Base_Dados.isnull().sum()
```

```python
Base_Dados.describe()
```

```python
plt.title('Analise dados nulos')
sns.heatmap(Base_Dados.isnull(), cbar=False);
```

```python
plt.figure(figsize=(10,6))
plt.title('Analise da Renda')
sns.kdeplot (Base_Dados['Renda'], color ='green');
```

```python
plt.title('Média de renda')
sns.boxplot(Base_Dados['Renda'])
```

```python
plt.title('Média de Experiência em anos')
sns.boxplot(Base_Dados['Xp'])
```

```python
sns.scatterplot(data=Base_Dados, x='Renda', y='Xp')
#Relação Linear, que comprova quanto maior a experiência, maior será a sua renda
```

```python
#Prova da Relação Linear
plt.figure(figsize=(12,7))
plt.title('Correlação entre Experiência no mercado e sua renda')
sns.regplot(data=Base_Dados, x='Renda', y='Xp')
```

```python
Correlacao = Base_Dados.corr('pearson')
sns.heatmap(Correlacao, annot=True, cbar=False)
```

```python
CorrelacaoS = Base_Dados.corr('spearman')
sns.heatmap(CorrelacaoS, annot=True, cbar=False)
```
