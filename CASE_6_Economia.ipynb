```python
# Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Base_Dados = pd.read_excel('Dados_Pib.xlsx')

Base_Dados.head()
```

```python
sns.heatmap(data=Base_Dados.isnull(), cbar=False)
```

```python
Base_Dados.groupby(by=['Territorialidades', 'Ano']).mean()
```

```python
#Sistema de GRIDS

#Cor de fundo
Cor_Fundo = '#f5f5f5'

#Criando o sistema de GRIDS

Grid_Graficos = sns.FacetGrid(Base_Dados, col='Territorialidades', hue='Territorialidades', col_wrap=4)


Grid_Grafico = Grid_Graficos.map(plt.plot, 'Ano', 'PIB per capita')

#SOMBRA + Titulo

Grid_Graficos = Grid_Graficos.map(plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name}')

#Adicionar Suptitulo
Grid_Graficos = Grid_Graficos.fig.suptitle('Projeto prático 6 - Análise PIB Per Capita no Brasil por Estado', fontsize=22, color='#404040', fontweight=600)

#AJUSTANDO SUPTITULO

plt.subplots_adjust(top=0.92)
```
