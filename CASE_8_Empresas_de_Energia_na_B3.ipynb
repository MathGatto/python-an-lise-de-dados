```python
# Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lendo a base de dados
BD = pd.read_excel('Dados_Empresas_Energia.xlsx')

# Verificando
BD.head()
```

```python
BD.set_index('Data', inplace=True)

BD.head()
```

```python
plt.style.use('seaborn-darkgrid');

plt.figure(figsize=(16,7));

plt.title('Anlise de aes de empresas de Energia', loc='left', fontsize=18, fontweight=600);

#PETROBRAS
plt.plot(BD.index,BD['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7)
plt.text(BD.index[-1], BD['Petrobras'].tail(1), ' Petrobras', color='#008c4a', size='large', horizontalalignment='left')


for Coluna in BD.columns[1:]:
  plt.plot(BD.index,BD[Coluna], color='gray', linewidth=2, alpha=0.7)
  plt.text(BD.index[-1], BD[Coluna].tail(1), Coluna, color='gray', size='large', horizontalalignment='left')


plt.xlabel('Perodo')
plt.ylabel('Valor de Fechamento (R$)');

```
